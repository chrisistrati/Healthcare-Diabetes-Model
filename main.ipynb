```python
#replacing abnormalities (values that shouldn't be 0 but are) with the mean value of the column

data["BMI"].replace(0, data["BMI"].mean())
data.replace({"SkinThickness": data["SkinThickness"].mean()}, inplace=True)
data.replace({"Glucose": data["Glucose"].mean()}, inplace=True)
data.replace({"BloodPressure": data["BloodPressure"].mean()}, inplace=True)
data.replace({"Insulin": data["Insulin"].mean()}, inplace=True)

data.sort_values("Insulin")
```
