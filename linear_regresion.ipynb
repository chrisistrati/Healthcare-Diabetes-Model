```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#imports
```

```python
data=pd.read_csv("Healthcare-Diabetes.csv")
data
#reading csv
```

```python
data.info()
```

```python
data.describe()
```

```python
#replacing abnormalities (values that shouldn't be 0 but are) with the mean value of the column

data["BMI"].replace(0, data["BMI"].mean())
data["SkinThickness"].replace({0: data["SkinThickness"].mean()}, inplace=True)
data["Glucose"].replace({0: data["Glucose"].mean()}, inplace=True)
data["BloodPressure"].replace({0: data["BloodPressure"].mean()}, inplace=True)
data["Insulin"].replace({0: data["Insulin"].mean()}, inplace=True)

data.sort_values("Insulin")
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data[data.columns[1:9]], data["Outcome"], test_size=0.1, random_state=0)
#splitting the data in testing and training data
```

```python
data[data.columns[2:8]]
```

```python
lr=LinearRegression() #creating the model
lr.fit(X_train, Y_train)
```

```python
y_pred_train=lr.predict(X_train) #predicting the answers for training data
y_pred_train
```

```python
from sklearn.metrics import r2_score
r2_score(Y_train, y_pred_train) #checking the score
```

```python
y_pred_test=lr.predict(X_test) #predicting aswers for testing data

r2_score(Y_test, y_pred_test) #checking the score

y_pred_test
```
